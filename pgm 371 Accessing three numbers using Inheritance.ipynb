```python
class A2nos():
    __x=None
    __y=None
    
    def __init__(self,x=500,y=350):
        self.__x=x
        self.__y=y
    def disp(self):
        print(self.__x,self.__y)
    def setx(self,x=23):
        self.__x=x
    def sety(self,y=78):
        self.__y=y
    def setxy(self,x=89,y=90):
        self.__x=x
        self.__y=y
    def setobj(self,ob):
        self.__x=ob.__x
        self.__y=ob.__y
    def reset(self):
        self.__x=500
        self.__y=350
    def getx(self):
        return self.__x
    def gety(self):
        return self.__y
    def getxy(self):
        return self.__x,self.__y
    def getobj(self):
        return self
```

```python
"""
pg.no.372
23/12/23
Accessing three numbers
"""

class A3nos(A2nos):
    __z=None
    def __init__(self,a=10,b=20,c=30):
        super().__init__(a,b)
        self.__z=c
    
    def displayz(self):
        print(self.__z)
        
    def displayall(self):
        m=self.getxy()
        print(m,self.__z)
        
    def setz(self,c):
        self.__z=c
        
    def setxyz(self,a=100,b=200,c=300):
        self.setxy(a,b)
        self.__z=c
    
    def reset(self):
        super().reset()
        self.__z=30
        
    def setobject(self,a):
        self.__z=a.__z
        
    def getz(self):
        return self.__z
    
    def getxyz(self):
        m=super().getxy()
        return(m,self.__z)
    
    def getobject(self):
        return self
        
        
#main
a=A3nos()
a.displayz()
a.displayall()

a.setz(500)
a.displayall()

a.setxyz(32,42,52)
a.displayall()

a.reset()
a.displayall()

obj=A3nos()
obj.setxyz(1001,2002,3003)
a.setobject(obj)
a.displayall()

x=a.getz()
print(x)

x=a.getobject()
x.setxyz(700,810,920)
print(x.getxyz())


```
