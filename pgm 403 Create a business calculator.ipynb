```python
"""
pg.no. 403 
13/01/2024 
Create a business calculator.
"""
import tkinter as tk

w=tk.Tk()
def mainwindow():
    w.title('Business calculator')
    w.configure(background='black')
    w.minsize(350,400)
    w.maxsize(1520,900)
    w.geometry("00x400+90+90" )
    w.attributes('-alpha',1)
    w.resizable(False, False)
    
global a
a=''

def add(v):
    global a
    a+=v
    l2.configure(text=a)
def design():
    f=tk.Frame(w, width=280, height=215, bg='grey', highlightbackground='red', highlightthickness=3)
    f.grid(row=20, column=10, padx=10, pady=10, ipadx=20, ipady=20)
    
    #labels
    global l2
    l2=tk.Label(w,text='',bg='white', fg='black', width=20,height=2, font=('times',20))
    l2.grid(row=10,column=10,padx=20, pady=20)
    
    #buttons
    b0=tk.Button(f,text='clear',height=1,width=4,bg='black',fg='white', font=('times',15),command=clear)
    b0.place(x=10, y=5)
    #b0.grid(row=15,column=10)
    
    b1=tk.Button(f,text='1',height=1,width=4, bg='black', fg='white', font=('times',15), command=one)
    b1.place(x=10, y=50)#grid(row=20, column=10)
    
    b2=tk.Button(f,text='2',height=1,width=4, bg='black', fg='white', font=('times',15), command=two)
    b2.place(x=90, y=50)#grid(row=20, column=17)
    
    b3=tk.Button(f,text='3',height=1,width=4, bg='black', fg='white', font=('times',15),command=three)
    b3.place(x=170, y=50) 
    
    b12=tk.Button(f,text='+',height=1,width=4,bg='black', fg='white', font=('times',15), command=plus)
    b12.place(x=240, y=50)
    
    b4=tk.Button(f,text='4',height=1,width=4, bg='black', fg='white', font=('times',15),command=four)
    b4.place(x=10 ,y=100) #grid(row=600+300, column=100)
    
    b5=tk.Button(f,text='5',height=1,width=4, bg='black', fg='white', font=('times',15),command=five)
    b5.place(x=90, y=100)#grid(row=600+300, column=140)
    
    b6=tk.Button(f,text='6',height=1,width=4, bg='black', fg='white', font=('times',15),command=six)
    b6.place(x=170,y=100) #grid(row=600+300, column=100)
    
    b13=tk.Button(f,text='-',height=1,width=4, bg='black', fg='white', font=('times',15), command=minus)
    b13.place(x=240, y=100) #grid(row=200+330, column=160)
    
    b7=tk.Button(f,text='7',height=1,width=4, bg='black', fg='white', font=('times',15),command=seven)
    b7.place(x=10, y=150)#grid(row=400+130,column=90)
    
    b8=tk.Button(f,text='8',height=1,width=4, bg='black', fg='white', font=('times',15),command=eight)
    b8.place(x=90, y=150) #grid(row=230+400, column=100)
    
    b9=tk.Button(f,text='9',height=1,width=4, bg='black', fg='white', font=('times',15),command=nine)
    b9.place(x=170, y=150) #grid(row=230+400, column=100)
    
    b14=tk.Button(f,text='*',height=1,width=4, bg='black', fg='white', font=('times',15), command=prod)
    b14.place(x=240, y=150) #grid(row=200+430, column=160)
    
    b10=tk.Button(f,text='0',height=1,width=4, bg='black', fg='white', font=('times',15),command=zero)
    b10.place(x=10, y=200)#grid(row=230+400, column=100)
    
    b11=tk.Button(f,text='.',height=1,width=4, bg='black', fg='white', font=('times',15),command=dot)
    b11.place(x=90, y=200)#grid(row=230+500, column=100)
    
    b15=tk.Button(f,text='/', bg='black', fg='white',height=1,width=4, font=('times',15), command=divide)
    b15.place(x=170, y=200) #grid(row=200+530, column=160)
    
    
    #equals
    b15=tk.Button(f,text='=',height=1,width=4, bg='black', fg='white', font=('times',15), command=equals)
    b15.place(x=240, y=200)#grid(row=9200,column=140)
    
    w.mainloop()
    
    
def one():
    global a
    a+='1'
    l2.configure(text=a)
    
def two():
    global a
    a+='2'
    l2.configure(text=a)
def three():
    global a
    a+='3'
    l2.configure(text=a)
def four():
    global a
    a+='4'
    l2.configure(text=a)
def five():
    global a
    a+='5'
    l2.configure(text=a)
def six():
    global a
    a+='6'
    l2.configure(text=a)
def seven():
    global a
    a+='7'
    l2.configure(text=a)
def eight():
    global a
    a+='8'
    l2.configure(text=a)
def nine():
    global a
    a+='9'
    l2.configure(text=a)
def zero():
    global a
    a+='0'
    l2.configure(text=a)
def dot():
    global a
    a+='.' 
    l2.configure(text=a)

def plus():
        global a
        a+='+'
        l2.configure(text=a)
        
def minus():
        global a
        a+='-' 
        l2.configure(text=a)
        
def prod():
        global a
        a+='*' 
        l2.configure(text=a)
        
def divide():
        global a
        a+='/' 
        l2.configure(text=a)

def equals():
    global a
    if a!='':
        try:
            result=eval(a)
        except:
            result="Error"
    l2.configure(text=result)
    
def clear():
    global a
    a=''
    l2.configure(text=a)
           
#main
mainwindow()
design()

    
    
```
