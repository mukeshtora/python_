```python
"""
pg.no.316
18/11/23
write the menu function, get the input number. if the input is 1 then print pattern 1
or if 2 print pattern 2 and so on
"""
def menu():
    print("enter '1' for single line of star pattern")
    print("enter '2' for n no. of stars in single line pattern ")
    print("enter '3' for n no. of stars in more than one rows")
    print("enter '4' for right triangle star pattern")
    print("enter '5' for mirror right triangle star pattern")
    print("enter '6' for right half-diamond pattern")
    print("enter '7' for upper traingle with hyphen")
    print("enter '8' for lower triangle wiht hyphen")
    print("enter '9' for upper & lower triangle with hyphen")
    print("enter '10' for upper & lower triangle")
    print("enter '11' for lower & upper traingle")
    print("enter '12' for pascal triangle")
    print("enter '13' for mirror pascal traingle")
    print("enter '14' for diamond star pattern")
    print("enter '15' for diaond number pattern")
    print("enter '16' for diamond alphabet pattern")
    print()
    print("enter '0' to stop")
    
    x=int(input('Enter number:'))
    return x

def pattern1():
    print('*'*6)
    
def pattern2(n):
    print('*'*n)
    
def pattern3():
    for i in range(5):
        print('*'*n)

def pattern4():
    for i in range(6):
        print('*'*i)
        
def pattern5():
    for i in range(5,0,-1):
        print('*'*i)
        
def pattern6():
    for i in range(6):
        print('*'*i)
    for i in range(4,0,-1):
        print('*'*i)

def pattern7():
    n=1
    for i in range(4,0,-1):
        print('-'*i+'*'*n)
        n=n+1
    print('*'*n)
    
def pattern8():
    n=5
    for i in range(5):
        print('-'*i+'*'*n)
        n=n-1
        
def pattern9():
    n=1
    for i in range(4,0,-1):
        print('-'*i+'*'*n)
        n=n+1
    n=5
    for i in range(0,5):
        print('-'*i+'*'*n)
        n=n-1
        
def pattern10():
    n=1
    for i in range(4,0,-1):
        print(' '*i+'*'*n)
        n=n+1
    n=5
    for i in range(5):
        print(' '*i+'*'*n)
        n=n-1
        
def pattern11():
    n=5
    for i in range(4):
        print(' '*i+'*'*n)
        n=n-1
    n=1
    for i in range(4,0,-1):
        print(' '*i+'*'*n)
        n=n+1
    
    print('*'*n)
    
def pattern12():
    n=1
    for i in range(8,0,-2):
        print(' '*i+'* '*n)
        n+=2
    
def pattern13():
    n=7
    for i in range(0,8,2):
        print(' '*i+'* '*n)
        n-=2
    
def pattern14():
    n=1
    for i in range(7,0,-2):
        print(' '*i+'* '*n)
        n+=2
    n=5
    for i in range(3,8,2):
        print(' '*i+'* '*n)
        n-=2

def pattern15():
    n=1
    x=int(input("enter any odd number within 9:"))
    for i in range(x,0,-2):
        for j in range(i):
            print(' ',end='')
        for k in range(1,n+1):
            print(k,end=' ')
        print()
        n+=2
    n=x
    m=3
    for i in range(x-2,0,-2):
        for j in range(m,0,-1):
            print(' ',end='')
        for k in range(1,n-1):
            print(k,end=' ')

        print()
        n-=2
        m+=2
    
def pattern16():
    n=1
    x=int(input("enter any odd number within 9:"))
    for i in range(x,0,-2):
        for j in range(i):
            print(' ',end='')
        for k in range(1,n+1):
            g=k+96
            y=chr(g)
            print(y,end=' ')
        print()
        n+=2
    n=x
    m=3
    for i in range(x-2,0,-2):
        for j in range(m,0,-1):
            print(' ',end='')
        for k in range(1,n-1):
            g=k+96
            y=chr(g)
            print(y,end=' ')
        print()
        n-=2
        m+=2

#Main

a=menu()
while a!=0:
    if a==1:
        pattern1()
    if a==2:
        n=int(input('enter n:'))
        pattern2(n)
    if a==3:
        pattern3()
    if a==4:
        pattern4()
    if a==5:
        pattern5()
    if a==6:
        pattern6()
    if a==7:
        pattern7()
    if a==8:
        pattern8()
    if a==9:
        pattern9()
    if a==10:
        pattern10()
    if a==11:
        pattern11()
    if a==12:
        pattern12()
    if a==13:
        pattern13()
    if a==14:
        pattern14()
    if a==15:
        pattern15()
    if a==16:
        pattern16()
    a=menu()
    
```
