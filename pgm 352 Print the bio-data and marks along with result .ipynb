```python
"""
pg.no.352
04/12/23
print the bio-data and marks along with result (pass/fail)
"""
import pickle
f=open('file12.dat','rb')
try:
    x=pickle.load(f)
    while x:
        print('Bishop Heber College')
        print(' '*7+'Trichy')
        print()
        
        k=list(x.keys())
        v=list(x.values())
        
        print('name:',v[0][0],' '*5,'rollno:',k[0])
        print('subject',' '*7,'marks',' ','Result')
        
        r='pass' if v[0][3]>=35 else 'fail'
        print('Tamil',' '*9,v[0][3],' '*5,r)
        
        r1='pass' if v[0][4]>=35 else 'fail'
        print('English',' '*7,v[0][4],' '*5,r1)
        
        r2='pass' if v[0][5]>=35 else 'fail'
        print('Maths',' '*9,v[0][5],' '*5,r2)
        
        r3='pass' if v[0][6]>=35 else 'fail'
        print('Physics',' '*7,v[0][6],' '*5,r3)
        
        r4='pass' if v[0][7]>=35 else 'fail'
        print('Social',' '*8,v[0][7],' '*5,r4)
        print()
        
        x=pickle.load(f)
        
except EOFError:
    print(' '*3,'*** End of File ***')
```
