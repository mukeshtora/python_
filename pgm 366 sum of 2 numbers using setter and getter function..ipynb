```python
"""
pg.no 366
14/12/23
sum of 2 numbers using setter and getter function.
"""
class sum2():
    __x=None
    __y=None
    __A=None
    
    def __init__(self):
        self.__x=10
        self.__y=20
        
    def find(self):
        self.__A=self.__x+self.__y
    def disp(self):
        print(self.__A)
        
    def setx(self,a=55):
        self.__x=a
    def sety(self,a=70):
        self.__y=a
    def setxy(self,a=56,b=90):
        self.__x=a
        self.__y=b
    def setobj(self,obj):
        self.__x=obj.__x
        self.__y=obj.__y
        
    def reset(self):
        self.__x=10
        self.__y=20
    def getx(self):
        return self.__x
    def gety(self):
        return(self.__y)
    def getxy(self):
        return(self.__x,self.__y)
    def getobj(self,a):
        return(self)
    
#main
a=sum2()
a.find()
a.disp()

a.setx()
a.sety()
a.find()
a.disp()

a.setxy()
a.find()
a.disp()

n=sum2()
n.setx(1000)
a.setobj(n)
a.find()
a.disp()

a.reset()
a.find()
a.disp()

b=a.getx()
print(b)
b=a.gety()
print(b)
b=a.getxy()
print(b)

b=a.getobj(n)
b.setobj(n)
print(b.getxy())

```
