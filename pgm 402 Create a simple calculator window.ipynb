### Using Config() method to alter the result(o/p) label.

```python
import tkinter as tk
w=tk.Tk()


def mainwindow():
    w.title('Login window')
    w.configure(background='skyblue')
    w.minsize(600,400)
    w.maxsize(1520,900)
    w.geometry("00x400+90+90")
    w.attributes('-alpha',1)
    
a=tk.IntVar()
b=tk.IntVar()
z=tk.IntVar()


def design():
    #labels
    l1=tk.Label(w,text='Enter A',bg='grey',fg='black',font=('times',15))
    l1.place(x=50,y=100,w=100,h=50)
    e1=tk.Entry(w,textvariable=a, bg='white',fg='black',font=('times',15))
    e1.place(x=250,y=100,w=200,h=50)
    l2=tk.Label(w,text='Enter B',bg='grey',fg='black',font=('times',15))
    l2.place(x=50,y=150+20,w=100,h=50)
    e2=tk.Entry(w,textvariable=b,bg='white',fg='black',font=('times',15))
    e2.place(x=250,y=150+20,w=200,h=50)
    l3=tk.Label(w,text='Result',bg='white',fg='black',font=('times',15))
    l3.place(x=50,y=240,w=100,h=50)
    #output label
    global l4
    l4=tk.Label(w,text='',bg='white',fg='black',font=('times',15))
    l4.place(x=250,y=240,w=100,h=50)
    
    #buttons
    b1=tk.Button(w,text='+',bg='grey',fg='black',font=('times',15),command=add)
    b1.place(x=150,y=310,w=40,h=40)
    b2=tk.Button(w,text='-',bg='grey',fg='black',font=('times',15),command=sub)
    b2.place(x=200,y=310,w=40,h=40)
    b3=tk.Button(w,text='*',bg='grey',fg='black',font=('times',15),command=prod)
    b3.place(x=250,y=310,w=40,h=40)
    b4=tk.Button(w,text='/',bg='grey',fg='black',font=('times',15),command=div)
    b4.place(x=300,y=310,w=40,h=40)
    w.mainloop()
    
def add():
    x=a.get()
    y=b.get()
    l4.config(text=x+y)
    
    
def sub():
    x=a.get()
    y=b.get() 
    l4.config(text=x-y)

def prod():
    x=a.get()
    y=b.get()
    l4.config(text=x*y)
    
def div():
    x=a.get()
    y=b.get()
    l4.config(text=x/y)

#main
mainwindow()
design()
```
