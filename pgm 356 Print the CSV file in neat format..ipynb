```python
"""
pg.no.356
07/12/23
Print the CSV file in neat format.
"""
import csv
f=open("filehandling(ls).csv")
x=csv.reader(f)

try:
    y=next(x)
    while x:
        print()
        print(" "*9,"Bishop Heber College")
        print(" "*15,"Trichy")
        y2=next(x)
        print("_____________________________________")
        print('|',y[0],':',y2[0],' '*7,y[1],':',y2[1],' |')
        print("|-----------------------------------|")
        print('|',y[2],':',y2[2]," "*12,'|')
        print("|-----------------------------------|")
        print('|',"subject"," "*5,'|',"mark"," "*3,'|',"result",'|')
        print("|___________________________________|")
        a="pass" if int(y2[3])>=35 else "fail"
        print('|',y[3],' '*7,'|',y2[3],' '*7,'|',a,'|')
        
        b="pass" if int(y2[4])>=35 else "fail"
        print('|',y[4],' '*5,'|',y2[4],' '*7,'|',b,'|')
        
        c="pass" if int(y2[5])>=35 else "fail"
        print('|',y[5]," "*7,'|',y2[5],((" "*6) if int(y2[5])>99 else (" "*7)),'|',c,'|')
        
        d="pass" if int(y2[6])>=35 else "fail"
        print('|',y[6]," "*5,'|',y2[6]," "*7,'|',d,'|')
        
        d="pass" if int(y2[7])>=35 else "fail"
        print('|',y[7]," "*6,'|',y2[7]," "*7,'|',d,'|')
        print("|___________________________________|")
        
        print("percentage  :",y2[8])
        print("final result:",y2[9])
        print(" "*13,"*** End ***")
        
except StopIteration:
        print()

        
```
