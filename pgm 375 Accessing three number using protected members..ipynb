```python
class A2nospro():
    _x=None
    _y=None
    
    def __init__(self,x=500,y=350):
        self._x=x
        self._y=y
    def disp(self):
        print(self._x,self._y)
    def setx(self,x=23):
        self._x=x
    def sety(self,y=78):
        self._y=y
    def setxy(self,x=89,y=90):
        self._x=x
        self._y=y
    def setobj(self,ob):
        self._x=ob._x
        self._y=ob._y
    def reset(self):
        self._x=500
        self._y=350
    def getx(self):
        return self._x
    def gety(self):
        return self._y
    def getxy(self):
        return self._x,self._y
    def getobj(self):
        return self
```

```python
"""
pg.no.375
23/12/23
Accessing three numbers using protected members.
"""
class A3nospro (A2nospro):
    _z=None
    
    def __init__(self,a=100,b=200,c=300):
        self._x=a
        self._y=b
        self._z=c
        
    def setz(self,a):
        self._z=a
        
    def setxyz(self,a,b,c):
        self._x=a
        self._y=b
        self._z=c
        
    def reset(self):
        self._x=100
        self._y=200
        self._z=300
        
    def setobject(self,a):
        self._x=a._x
        self._y=a._y
        self._z=a._z
        
    def displayz(self):
        print(self._z)
        
    def displayall(self):
        print(self._x,self._y,self._z)
        
#main
a=A3nospro()
a.displayz()
a.displayall()

a.setz(90)
a.displayall()

a.setxyz(45,90,180)
a.displayall()

a.reset()
a.displayall()

n=A3nospro()
n.setz(180)

a.setobject(n)
a.displayall()

```
